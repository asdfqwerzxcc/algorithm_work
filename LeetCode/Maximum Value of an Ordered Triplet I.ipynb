# Maximum Value of an Ordered Triplet I

You are given a 0-indexed integer array nums.

Return the maximum value over all triplets of indices (i, j, k) such that i < j < k. If all such triplets have a negative value, return 0.

The value of a triplet of indices (i, j, k) is equal to (nums[i] - nums[j]) * nums[k].

```python
class Solution:
    def maximumTripletValue(self, nums: List[int]) -> int:
        n = len(nums)
        if n < 3:
            return 0

        # prefix arrays: max_left와 min_left
        max_left = [0] * n
        min_left = [0] * n
        max_left[0] = nums[0]
        min_left[0] = nums[0]
        for i in range(1, n):
            max_left[i] = max(max_left[i-1], nums[i])
            min_left[i] = min(min_left[i-1], nums[i])
        
        # suffix arrays: j 이후의 양수 중 최대값과 음수 중 최소값을 구하기
        suffix_max = [None] * n  # 양수들 중 최댓값 (존재하지 않으면 None)
        suffix_min = [None] * n  # 음수들 중 최솟값 (존재하지 않으면 None)
        
        # 마지막 원소는 j 이후에 값이 없으므로 None 처리
        suffix_max[n-1] = None
        suffix_min[n-1] = None
        
        # 오른쪽부터 채우기 (인덱스 n-2 부터 0까지)
        current_max = None
        current_min = None
        for i in range(n-1, -1, -1):
            # 현재 값이 양수면 current_max 갱신
            if nums[i] >= 0:
                if current_max is None:
                    current_max = nums[i]
                else:
                    current_max = max(current_max, nums[i])
            # 현재 값이 음수면 current_min 갱신
            if nums[i] < 0:
                if current_min is None:
                    current_min = nums[i]
                else:
                    current_min = min(current_min, nums[i])
            suffix_max[i] = current_max
            suffix_min[i] = current_min

        max_value = float('-inf')
        
        # j는 1부터 n-2까지 선택 (i < j < k)
        for j in range(1, n-1):
            # 후보 1: k가 양수인 경우
            if max_left[j-1] > nums[j] and suffix_max[j+1] is not None:
                cand1 = (max_left[j-1] - nums[j]) * suffix_max[j+1]
                max_value = max(max_value, cand1)
            
            # 후보 2: k가 음수인 경우, (min_left[j-1] - nums[j])가 음수여야 함.
            if min_left[j-1] < nums[j] and suffix_min[j+1] is not None:
                diff = min_left[j-1] - nums[j]
                if diff < 0:  # 음수여야 음수*k에서 양수를 얻음
                    cand2 = diff * suffix_min[j+1]
                    max_value = max(max_value, cand2)
        
        return max_value if max_value >= 0 else 0
```
